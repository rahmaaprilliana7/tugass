 04 – Set dan Operasi Himpunan

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
```

 05 – Studi Kasus: Data Mahasiswa dan Mata Kuliah

```python
mhs1 = {"Algoritma", "Matematika", "Bahasa Inggris"}
mhs2 = {"Algoritma", "Basis Data", "Matematika"}

print("Mata kuliah Mhs1:", mhs1)
print("Mata kuliah Mhs2:", mhs2)
print("Mata kuliah yang sama:", mhs1 & mhs2)
```

06 – Studi Kasus Dictionary: Daftar Nilai Mahasiswa

```python
nilai_mahasiswa = {
    "Andi": 80,
    "Budi": 75,
    "Cici": 90
 }

for nama, nilai in nilai_mahasiswa.items():
    print(f"{nama}: {nilai}")

rata = sum(nilai_mahasiswa.values()) / len(nilai_mahasiswa)
print(f"Rata-rata nilai: {rata:.2f}")
```

8.Studi Kasus Dictionary

```python
nilai_mahasiswa = {"Andi": 85, "Budi": 90, "Cici": 78}

for nama,nilai in nilai_mahasiswa.items():
  print(f"{nama} : {nilai}")

  rata = sum(nilai_mahasiswa.values())/len(nilai_mahasiswa)
  print(f"\nrata-rata nilai : {rata:.2f}")
```


9.Studi Kasus Set – Klub Mahasiswa

```python
Klub_A = {"Andi", "Budi", "Cici"}
Klub_B = {"Cici", "Deni", "Eka"}

print("Daftar seluruh anggota", Klub_A | Klub_B)
print("Anggota klub A saja:", Klub_A - Klub_B)
print("anggota yang ada di kedua klub:", Klub_A - Klub_B)
```
