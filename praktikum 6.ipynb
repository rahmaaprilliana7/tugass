 05 – Operasi pada List 2D 05 – Operasi pada List 2D

```python
A = [[1, 2, 3], [4, 5, 6]]
B = [[7, 8, 9], [10, 11, 12]]
C = []

for i in range(len(A)):
  baris = []
  for j in range(len(A[0])):
      baris.append(A[i][j] + B[i][j])
  C.append(baris)
print(C)
```

06 – Studi Kasus: Pengelolaan Nilai Mahasiswa

```python
nilai = [
  [80, 90, 85],
  [70, 75, 80],
  [88, 92, 84]
 ]

for i, baris in enumerate(nilai, start=1):
 rata = sum(baris) / len(baris)
 print(f"Rata-rata mahasiswa {i}: {rata:.2f}")

```

7.Operasi Matriks 2D

```python
A = [
  [1, 2],
  [3, 4]
]

B = [
  [5, 6],
  [7, 8]
]

hasil = [
  [A[0][0] + B[0][0], A[0][1] + B[0][1]],
  [A[1][0] + B[1][0], A[1][1] + B[1][1]]
]

print("Hasil penjumlahan matriks:")
for baris in hasil:
  print(baris)
```

9.Studi Kasus – Data Toko

```python
stok = [["Buku", 20], ["Pulpen", 35], ["Penghapus", 15]]

for nama, jumlah in (stok):
  print(f"{nama}: {jumlah}")

terbanyak = max(stok, key=lambda x: x[1])
print(f"\n stok terbanyak", terbanyak[0], "(",terbanyak[1], "unit)")
```
