6. String Formatting dengan F-String

```python
data = []

nama = input('masukkan nama: ')
umur = input('masukkan umur: ')
jurusan = input('masukkan jurusan: ')

#memasukkan input ke dalam data
data.append({
    'nama': nama,
    'umur': umur,
    'jurusan': jurusan
})

#menampilkan data
for i in data:
  print(f'\nNama: {i["nama"]}')
  print(f'Umur: {i["umur"]}')
  print(f'Jurusan: {i["jurusan"]}')
```

7. Operasi dalam F-String

```python
a = int(input('masukkan bilangan a: '))
b = int(input('masukkan bilangan b: '))

print(f'hasil penjumlahan {a}+{b} = {a+b}')
print(f'hasil perkalian {a}x{b} = {a*b}')
print(f'hasil pembagian {a} ÷ {b} = {a/b:.3f}')
```

8. Casting Data

```python
angka_str =input('masukkan angka:')

angka_int=int(angka_str)
kuadrat = angka_int ** 2

angka_float = float(angka_str)
angka_koma = f'{angka_float:.2f}'

print(f'kuadrat dari {angka_str} adalah {kuadrat}')
print(f'dalam format float adalah {angka_koma}')

```

9. Studi Kasus – Analisis Teks

```python
kalimat = input('masukkan kalimat: ')

daftar_kata = kalimat.split()
jumlah_kata = len(daftar_kata)

kata_terpanjang = max(daftar_kata, key=len)

kata_urut = sorted(daftar_kata)

print(f'\njumlah kata: {jumlah_kata}')
print(f'kata terpanjang: {kata_terpanjang}')
for kata in kata_urut:
  print(f'- {kata}')



```
